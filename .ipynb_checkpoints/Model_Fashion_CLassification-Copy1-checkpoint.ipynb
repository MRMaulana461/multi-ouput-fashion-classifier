```python
# Definisikan fungsi residual block
def res_block(x, filters, kernel_size=3, stride=1, use_bias=True, name=None):
    shortcut = x
    
    # First convolution
    x = Conv2D(filters, kernel_size, strides=stride, padding='same', use_bias=use_bias, 
               kernel_regularizer=l2(0.001), name=f'{name}_conv1')(x)
    x = BatchNormalization(name=f'{name}_bn1')(x)
    x = Activation('relu', name=f'{name}_relu1')(x)
    
    # Second convolution
    x = Conv2D(filters, kernel_size, padding='same', use_bias=use_bias, 
               kernel_regularizer=l2(0.001), name=f'{name}_conv2')(x)
    x = BatchNormalization(name=f'{name}_bn2')(x)
    
    # Shortcut connection
    if stride > 1 or shortcut.shape[-1] != filters:
        shortcut = Conv2D(filters, 1, strides=stride, padding='same', use_bias=use_bias,
                          kernel_regularizer=l2(0.001), name=f'{name}_shortcut')(shortcut)
        shortcut = BatchNormalization(name=f'{name}_shortcut_bn')(shortcut)
    
    # Add shortcut to output
    x = Add(name=f'{name}_add')([x, shortcut])
    x = Activation('relu', name=f'{name}_relu2')(x)
    
    return x
```

```python
# Encode targets
le_gender = LabelEncoder()
le_usage = LabelEncoder()
le_article = LabelEncoder()

df_merged['gender_encoded'] = le_gender.fit_transform(df_merged['gender'])
df_merged['usage_encoded'] = le_usage.fit_transform(df_merged['usage'])
df_merged['article_encoded'] = le_article.fit_transform(df_merged['articleType'])

print(f"Number of gender classes: {len(le_gender.classes_)}")
print(f"Number of usage classes: {len(le_usage.classes_)}")
print(f"Number of article classes: {len(le_article.classes_)}")

# Split data - stratified split for better class balance
train_df, val_df = train_test_split(
    df_merged,
    test_size=0.2,
    stratify=df_merged['gender_encoded'].astype(str) + '_' + 
             df_merged['usage_encoded'].astype(str) + '_' + 
             df_merged['article_encoded'].astype(str),
    random_state=42
)
```

```python
# Definisi helper function untuk multi_output_generator
def multi_output_generator(generator):
    while True:
        batch = next(generator)
        images, labels = batch
        # Ensure labels are numeric
        gender = labels[:, 0].astype(np.int32)
        usage = labels[:, 1].astype(np.int32)
        article = labels[:, 2].astype(np.int32)
        yield images, {'gender': gender, 'usage': usage, 'article': article}
```

```python
# Augmentasi data yang ditingkatkan untuk menangani variasi pencahayaan dan kualitas gambar
def create_robust_augmentation_generators():
    def robust_preprocessing(img):
        # Normalisasi standar
        img = (img - np.mean(img)) / (np.std(img) + 1e-7)
        
        # Simulasi gambar berkualitas rendah secara acak
        if np.random.random() < 0.3:  # 30% kemungkinan
            # Tambahkan noise
            noise_level = np.random.uniform(0.01, 0.05)
            noise = np.random.normal(0, noise_level, img.shape)
            img = img + noise
            img = np.clip(img, 0, 1)  # Pastikan nilai tetap valid
            
            # Blur acak
            if np.random.random() < 0.5:
                blur_size = np.random.choice([3, 5])
                img = cv2.GaussianBlur(img, (blur_size, blur_size), 0)
                
            # Kompresi JPEG simulasi
            if np.random.random() < 0.4:
                encode_param = [int(cv2.IMWRITE_JPEG_QUALITY), np.random.randint(40, 90)]
                _, encimg = cv2.imencode('.jpg', (img * 255.0).astype(np.uint8), encode_param)
                img = cv2.imdecode(encimg, 1) / 255.0
        
        # Simulasi variasi pencahayaan
        if np.random.random() < 0.4:  # 40% kemungkinan
            # Kontras acak
            alpha = np.random.uniform(0.6, 1.4)
            img = img * alpha
            
            # Kecerahan acak per channel (simulasi white balance yang berbeda)
            beta = np.random.uniform(0.7, 1.3, size=(1, 1, 3))
            img = img * beta
            
            # Gamma koreksi untuk simulasi kondisi gelap/terang
            gamma = np.random.uniform(0.7, 1.3)
            img = np.power(np.maximum(img, 1e-10), gamma)  # Hindari nilai negatif pada power
        
        # Color jitter (mengubah saturasi dan hue)
        if np.random.random() < 0.3:
            # Pastikan gambar dalam range yang benar dan konversi ke tipe data float32
            img_rgb = np.clip(img, 0, 1).astype(np.float32)
            # Konversi ke HSV
            hsv = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2HSV)
            # Ubah hue
            hsv[:,:,0] = hsv[:,:,0] * np.random.uniform(0.8, 1.2)
            # Ubah saturasi
            hsv[:,:,1] = hsv[:,:,1] * np.random.uniform(0.6, 1.4)
            # Konversi kembali ke RGB
            img = cv2.cvtColor(hsv, cv2.COLOR_HSV2RGB)
            
        # Random occlusion/erasing - simulasi objek tertutup sebagian
        if np.random.random() < 0.2:
            h, w = img.shape[0], img.shape[1]
            area = h * w
            target_area = np.random.uniform(0.02, 0.15) * area
            aspect_ratio = np.random.uniform(0.3, 1/0.3)
            
            # Hitung tinggi dan lebar
            h_occlusion = int(round(np.sqrt(target_area * aspect_ratio)))
            w_occlusion = int(round(np.sqrt(target_area / aspect_ratio)))
            
            if h_occlusion < h and w_occlusion < w:
                x1 = np.random.randint(0, w - w_occlusion)
                y1 = np.random.randint(0, h - h_occlusion)
                
                if len(img.shape) == 3:
                    img[y1:y1+h_occlusion, x1:x1+w_occlusion, :] = np.random.uniform(0, 1, size=(h_occlusion, w_occlusion, img.shape[2]))
                else:
                    img[y1:y1+h_occlusion, x1:x1+w_occlusion] = np.random.uniform(0, 1)
        
        # Pastikan nilai valid
        img = np.clip(img, 0, 1.0)
        return img

    # Generator untuk training dengan augmentasi yang lebih robust
    train_datagen = ImageDataGenerator(
        rescale=1./255,
        rotation_range=20,                # Rotasi moderat untuk pakaian
        zoom_range=0.2,                   # Zoom untuk variasi skala
        width_shift_range=0.2,            # Pergeseran horizontal
        height_shift_range=0.2,           # Pergeseran vertikal
        horizontal_flip=True,             # Flip horizontal
        fill_mode='nearest',
        preprocessing_function=robust_preprocessing
    )

    # Generator untuk validasi dengan normalisasi dasar
    val_datagen = ImageDataGenerator(
        rescale=1./255,
        preprocessing_function=lambda x: (x - np.mean(x)) / (np.std(x) + 1e-7)  # Hanya normalisasi
    )

    return train_datagen, val_datagen
```

```python
# Fungsi untuk menerapkan robust augmentation generator
def apply_robust_augmentation():
    # Buat generator dengan augmentasi yang lebih robust
    robust_train_datagen, robust_val_datagen = create_robust_augmentation_generators()
    
    # Konfigurasi generator untuk training
    robust_train_generator_raw = robust_train_datagen.flow_from_dataframe(
        dataframe=train_df,
        x_col='image_path',
        y_col=['gender_encoded', 'usage_encoded', 'article_encoded'],
        target_size=IMG_SIZE,
        batch_size=BATCH_SIZE,
        class_mode='raw',
        shuffle=True
    )
    
    # Konfigurasi generator untuk validasi
    robust_val_generator_raw = robust_val_datagen.flow_from_dataframe(
        dataframe=val_df,
        x_col='image_path',
        y_col=['gender_encoded', 'usage_encoded', 'article_encoded'],
        target_size=IMG_SIZE,
        batch_size=BATCH_SIZE,
        class_mode='raw',
        shuffle=False
    )
    
    # Wrap generator untuk memastikan format output yang tepat
    robust_train_generator = multi_output_generator(robust_train_generator_raw)
    robust_val_generator = multi_output_generator(robust_val_generator_raw)
    
    return robust_train_generator, robust_val_generator
```

```python
# Modifikasi pada fungsi build_improved_cnn untuk meningkatkan ketahanan terhadap variasi gambar
def build_robust_cnn(input_shape, n_gender, n_usage, n_article):
    inputs = Input(shape=input_shape)
    
    # Lapisan normalisasi di awal untuk menyeragamkan input
    x = Lambda(lambda x: (x - tf.reduce_mean(x, axis=[1, 2], keepdims=True)) / 
              (tf.math.reduce_std(x, axis=[1, 2], keepdims=True) + 1e-7))(inputs)
    
    # Konvolusi awal dengan filter lebih banyak untuk menangkap fitur lokal
    x = Conv2D(64, (5, 5), strides=(2, 2), padding='same', kernel_regularizer=l2(0.001))(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = MaxPooling2D(pool_size=(2, 2))(x)
    
    # Residual blocks dengan BatchNormalization (lebih stabil untuk batch kecil)
    x = res_block(x, 128, name='res_block1')
    x = MaxPooling2D(pool_size=(2, 2))(x)
    x = SpatialDropout2D(0.25)(x)  # Spatial dropout lebih efektif untuk konvolusi
    
    x = res_block(x, 256, name='res_block2')
    x = MaxPooling2D(pool_size=(2, 2))(x)
    x = SpatialDropout2D(0.35)(x)
    
    x = res_block(x, 512, name='res_block3')
    x = MaxPooling2D(pool_size=(2, 2))(x)
    x = SpatialDropout2D(0.4)(x)
    
    # Feature Pyramid Network style - menggabungkan fitur dari beberapa skala
    # untuk menangkap detail pada berbagai tingkatan
    x1 = GlobalAveragePooling2D()(x)
    x2 = GlobalMaxPooling2D()(x)  # Max pooling untuk menangkap fitur paling menonjol
    x = Concatenate()([x1, x2])
    
    # Fully connected dengan dropout yang lebih tinggi
    x = Dense(512, kernel_regularizer=l2(0.001))(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = Dropout(0.5)(x)
    
    # Task-specific branches
    gender_branch = Dense(128, activation='relu', kernel_regularizer=l2(0.001))(x)
    gender_branch = BatchNormalization()(gender_branch)
    gender_branch = Dropout(0.4)(gender_branch)
    gender_output = Dense(n_gender, activation='softmax', name='gender')(gender_branch)
    
    usage_branch = Dense(128, activation='relu', kernel_regularizer=l2(0.001))(x)
    usage_branch = BatchNormalization()(usage_branch)
    usage_branch = Dropout(0.4)(usage_branch)
    usage_output = Dense(n_usage, activation='softmax', name='usage')(usage_branch)
    
    article_branch = Dense(256, activation='relu', kernel_regularizer=l2(0.001))(x)
    article_branch = BatchNormalization()(article_branch)
    article_branch = Dropout(0.5)(article_branch)
    article_output = Dense(n_article, activation='softmax', name='article')(article_branch)
    
    model = Model(inputs=inputs, outputs=[gender_output, usage_output, article_output])
    return model
```

```python
# Fungsi untuk melatih model dengan augmentasi robust
def train_with_robust_augmentation():
    # Buat model yang robust
    robust_model = build_robust_cnn(
        (IMG_SIZE[0], IMG_SIZE[1], 3),
        n_gender=len(le_gender.classes_),
        n_usage=len(le_usage.classes_),
        n_article=len(le_article.classes_)
    )
    
    # Generator dengan augmentasi robust
    robust_train_generator, robust_val_generator = apply_robust_augmentation()
    
    # Compile model dengan optimizer yang lebih robust
    optimizer = Adam(learning_rate=1e-3, clipnorm=1.0)  # Clip gradien untuk stabilitas
    
    robust_model.compile(
        optimizer=optimizer,
        loss={
            'gender': 'sparse_categorical_crossentropy',
            'usage': 'sparse_categorical_crossentropy',
            'article': 'sparse_categorical_crossentropy'
        },
        metrics=['accuracy'],
        loss_weights={
            'gender': 2.0, 
            'usage': 1.0, 
            'article': 2.0
        }
    )
    
    # Callbacks untuk mencegah overfitting
    robust_callbacks = [
        ModelCheckpoint(
            'best_robust_model.h5', 
            monitor='val_loss', 
            save_best_only=True, 
            mode='min', 
            verbose=1
        ),
        EarlyStopping(
            monitor='val_loss', 
            patience=15,  # Lebih sabar
            restore_best_weights=True, 
            verbose=1,
            min_delta=0.005
        ),
        ReduceLROnPlateau(
            monitor='val_loss', 
            factor=0.5, 
            patience=5,
            min_lr=1e-6, 
            verbose=1
        )
    ]
    
    # Membagi epochs training menjadi dua tahap
    print("\n===== TRAINING ROBUST CNN MODEL DENGAN AUGMENTASI LANJUTAN =====")
    steps_per_epoch = len(train_df) // BATCH_SIZE
    validation_steps = len(val_df) // BATCH_SIZE
    
    # Tahap 1: Latih model dengan learning rate yang tinggi
    print("\n--- Tahap 1: Initial Training ---")
    initial_history = robust_model.fit(
        robust_train_generator,
        steps_per_epoch=steps_per_epoch,
        validation_data=robust_val_generator,
        validation_steps=validation_steps,
        epochs=30,
        callbacks=robust_callbacks,
        verbose=1
    )
    
    # Tahap 2: Fine-tuning dengan learning rate lebih rendah
    print("\n--- Tahap 2: Fine Tuning ---")
    # Atur ulang optimizer dengan learning rate lebih rendah
    K = tf.keras.backend
    K.set_value(robust_model.optimizer.learning_rate, 1e-4)
    
    fine_tune_history = robust_model.fit(
        robust_train_generator,
        steps_per_epoch=steps_per_epoch,
        validation_data=robust_val_generator,
        validation_steps=validation_steps,
        epochs=70,  # Tambahan epoch untuk fine-tuning
        callbacks=robust_callbacks,
        verbose=1
    )
    
    # Gabungkan history
    combined_history = {
        'loss': initial_history.history['loss'] + fine_tune_history.history['loss'],
        'val_loss': initial_history.history['val_loss'] + fine_tune_history.history['val_loss'],
        'gender_accuracy': initial_history.history['gender_accuracy'] + fine_tune_history.history['gender_accuracy'],
        'val_gender_accuracy': initial_history.history['val_gender_accuracy'] + fine_tune_history.history['val_gender_accuracy'],
        'usage_accuracy': initial_history.history['usage_accuracy'] + fine_tune_history.history['usage_accuracy'],
        'val_usage_accuracy': initial_history.history['val_usage_accuracy'] + fine_tune_history.history['val_usage_accuracy'],
        'article_accuracy': initial_history.history['article_accuracy'] + fine_tune_history.history['article_accuracy'],
        'val_article_accuracy': initial_history.history['val_article_accuracy'] + fine_tune_history.history['val_article_accuracy']
    }
    
    return robust_model, combined_history
```

```python
# Fungsi untuk evaluasi model pada gambar yang berkualitas rendah
def evaluate_on_degraded_images(model, val_generator, val_steps):
    # Generator untuk evaluasi dengan simulasi gambar berkualitas rendah
    def degrade_image(img):
        # Normalisasi standar
        img = (img - np.mean(img)) / (np.std(img) + 1e-7)
        
        # Ciptakan kondisi ekstrem (lebih dari yang dilatih)
        # Tambahkan noise tingkat tinggi
        noise = np.random.normal(0, 0.1, img.shape)
        img = img + noise
        
        # Tambahkan blur yang kuat
        img = cv2.GaussianBlur(img, (7, 7), 0)
        
        # Simulasi pencahayaan yang buruk
        brightness_factor = np.random.uniform(0.4, 0.7)
        img = img * brightness_factor
        
        # Simulasi kualitas kompresi rendah
        encode_param = [int(cv2.IMWRITE_JPEG_QUALITY), 30]  # Kualitas sangat rendah
        img_uint8 = (np.clip(img, 0, 1) * 255).astype(np.uint8)
        _, encimg = cv2.imencode('.jpg', img_uint8, encode_param)
        img = cv2.imdecode(encimg, 1) / 255.0
        
        # Normalisasi kembali
        img = (img - np.mean(img)) / (np.std(img) + 1e-7)
        img = np.clip(img, 0, 1.0)
        return img
    
    # Buat generator untuk evaluasi dengan gambar berkualitas rendah
    eval_datagen = ImageDataGenerator(
        rescale=1./255,
        preprocessing_function=degrade_image
    )
    
    # Konfigurasi generator evaluasi
    eval_generator_raw = eval_datagen.flow_from_dataframe(
        dataframe=val_df,
        x_col='image_path',
        y_col=['gender_encoded', 'usage_encoded', 'article_encoded'],
        target_size=IMG_SIZE,
        batch_size=BATCH_SIZE,
        class_mode='raw',
        shuffle=False
    )
    
    eval_generator = multi_output_generator(eval_generator_raw)
    
    # Evaluasi model
    print("\n===== EVALUASI MODEL PADA GAMBAR BERKUALITAS RENDAH =====")
    results = model.evaluate(
        eval_generator,
        steps=val_steps,
        verbose=1
    )
    
    # Cetak hasil evaluasi
    print(f"\nLoss: {results[0]}")
    print(f"Gender Accuracy: {results[4] * 100:.2f}%")
    print(f"Usage Accuracy: {results[5] * 100:.2f}%")
    print(f"Article Accuracy: {results[6] * 100:.2f}%")
    
    return results
```

```python
# Fungsi untuk visualisasi augmentasi
def visualize_augmentations():
    # Ambil beberapa gambar sampel
    sample_images = []
    sample_labels = []
    
    # Ambil beberapa sampel dari validation set
    val_generator_vis = ImageDataGenerator(rescale=1./255).flow_from_dataframe(
        dataframe=val_df.sample(5),
        x_col='image_path',
        y_col=['gender_encoded', 'usage_encoded', 'article_encoded'],
        target_size=IMG_SIZE,
        batch_size=5,
        class_mode='raw',
        shuffle=False
    )
    
    # Ambil batch pertama
    images, labels = next(val_generator_vis)
    
    # Dapatkan fungsi preprocessing robust
    _, _ = create_robust_augmentation_generators()
    
    # Fungsi untuk mengambil fungsi preprocessing dari train_datagen
    def get_preprocessing_function():
        def robust_preprocessing(img):
            # Normalisasi standar
            img = (img - np.mean(img)) / (np.std(img) + 1e-7)
            
            # Simulasi gambar berkualitas rendah secara acak
            if True:  # Selalu lakukan untuk visualisasi
                # Tambahkan noise
                noise_level = np.random.uniform(0.01, 0.05)
                noise = np.random.normal(0, noise_level, img.shape)
                img = img + noise
                img = np.clip(img, 0, 1)  # Pastikan nilai tetap valid
                
                # Blur acak
                if True:
                    blur_size = np.random.choice([3, 5])
                    img = cv2.GaussianBlur(img, (blur_size, blur_size), 0)
                    
                # Kompresi JPEG simulasi
                if True:
                    encode_param = [int(cv2.IMWRITE_JPEG_QUALITY), np.random.randint(40, 90)]
                    img_uint8 = (np.clip(img, 0, 1) * 255).astype(np.uint8)
                    _, encimg = cv2.imencode('.jpg', img_uint8, encode_param)
                    img = cv2.imdecode(encimg, 1) / 255.0
            
            # Simulasi variasi pencahayaan
            if True:  # Selalu lakukan untuk visualisasi
                # Kontras acak
                alpha = np.random.uniform(0.6, 1.4)
                img = img * alpha
                
                # Kecerahan acak per channel (simulasi white balance yang berbeda)
                beta = np.random.uniform(0.7, 1.3, size=(1, 1, 3))
                img = img * beta
                
                # Gamma koreksi untuk simulasi kondisi gelap/terang
                gamma = np.random.uniform(0.7, 1.3)
                img = np.power(np.clip(img, 1e-10, 1.0), gamma)
            
            # Pastikan nilai valid
            img = np.clip(img, 0, 1.0)
            return img
        
        return robust_preprocessing
    
    # Dapatkan fungsi preprocessing
    robust_preprocess = get_preprocessing_function()
    
    # Plot original dan augmented images
    plt.figure(figsize=(15, 10))
    
    for i in range(5):
        # Original image
        plt.subplot(5, 3, i*3 + 1)
        plt.imshow(images[i])
        plt.title(f'Original {i+1}')
        plt.axis('off')
        
        # Augmented image 1
        aug_img1 = robust_preprocess(images[i].copy())
        plt.subplot(5, 3, i*3 + 2)
        plt.imshow(np.clip(aug_img1, 0, 1))
        plt.title(f'Augmented 1 {i+1}')
        plt.axis('off')
        
        # Augmented image 2
        aug_img2 = robust_preprocess(images[i].copy())
        plt.subplot(5, 3, i*3 + 3)
        plt.imshow(np.clip(aug_img2, 0, 1))
        plt.title(f'Augmented 2 {i+1}')
        plt.axis('off')
    
    plt.tight_layout()
    plt.savefig('augmentation_visualization.png')
    plt.show()
```

```python
# Fungsi untuk melakukan prediksi dan evaluasi dengan confusion matrix
def evaluate_model_with_confusion_matrix(model, test_df, batch_size=32):
    # Generator untuk data evaluasi
    test_datagen = ImageDataGenerator(
        rescale=1./255,
        preprocessing_function=lambda x: (x - np.mean(x)) / (np.std(x) + 1e-7)
    )
    
    # Konfigurasi generator untuk testing
    test_generator = test_datagen.flow_from_dataframe(
        dataframe=test_df,
        x_col='image_path',
        y_col=['gender_encoded', 'usage_encoded', 'article_encoded'],
        target_size=(256, 256),  # IMG_SIZE dari kode asli
        batch_size=batch_size,
        class_mode='raw',
        shuffle=False
    )
    
    # Prediksi
    print("Melakukan prediksi pada set validasi...")
    steps = len(test_df) // batch_size + (1 if len(test_df) % batch_size > 0 else 0)
    
    # Mendapatkan label sebenarnya
    y_true_gender = test_df['gender_encoded'].values
    y_true_usage = test_df['usage_encoded'].values
    y_true_article = test_df['article_encoded'].values
    
    # Mendapatkan prediksi
    predictions = model.predict(test_generator, steps=steps)
    y_pred_gender = np.argmax(predictions[0], axis=1)
    y_pred_usage = np.argmax(predictions[1], axis=1)
    y_pred_article = np.argmax(predictions[2], axis=1)
    
    # Ambil jumlah sampel yang sesuai dengan hasil prediksi (jika batch tidak pas)
    y_true_gender = y_true_gender[:len(y_pred_gender)]
    y_true_usage = y_true_usage[:len(y_pred_usage)]
    y_true_article = y_true_article[:len(y_pred_article)]
    
    # Confusion Matrix untuk Gender
    cm_gender = confusion_matrix(y_true_gender, y_pred_gender)
    gender_labels = le_gender.classes_
    
    # Confusion Matrix untuk Usage
    cm_usage = confusion_matrix(y_true_usage, y_pred_usage)
    usage_labels = le_usage.classes_
    
    # Confusion Matrix untuk Article Type
    cm_article = confusion_matrix(y_true_article, y_pred_article)
    article_labels = le_article.classes_
    
    # Visualisasi Confusion Matrix
    plt.figure(figsize=(15, 12))
    
    plt.subplot(2, 2, 1)
    sns.heatmap(cm_gender, annot=True, fmt='d', cmap='Blues', xticklabels=gender_labels, yticklabels=gender_labels)
    plt.title('Confusion Matrix - Gender')
    plt.xlabel('Predicted')
    plt.ylabel('True')
    
    plt.subplot(2, 2, 2)
    sns.heatmap(cm_usage, annot=True, fmt='d', cmap='Blues', xticklabels=usage_labels, yticklabels=usage_labels)
    plt.title('Confusion Matrix - Usage')
    plt.xlabel('Predicted')
    plt.ylabel('True')
    
    # Untuk artikel, mungkin terlalu banyak kelas, kita fokus pada top misclassifications
    plt.subplot(2, 1, 2)
    
    # Jika article labels terlalu banyak, ambil subset
    if len(article_labels) > 15:
        # Fokus pada kelas yang bermasalah (heels, shirts, tshirts)
        problem_classes = ['Heels', 'Shirts', 'Tshirts', 'Sport Shoes', 'Casual Shoes']
        
        # Find indices of these classes
        problem_indices = [i for i, label in enumerate(article_labels) if any(pc in label for pc in problem_classes)]
        
        # If too few matches, use top 10 most commonly confused
        if len(problem_indices) < 5:
            # Ambil dari 10 kelas yang paling sering salah prediksi
            diag_indices = np.diag_indices(len(cm_article))
            diag_values = cm_article[diag_indices]
            off_diag_total = np.sum(cm_article, axis=1) - diag_values
            top_confused_idx = np.argsort(off_diag_total)[-10:]
            problem_indices = top_confused_idx
        
        # Extract the sub confusion matrix
        sub_cm = cm_article[np.ix_(problem_indices, problem_indices)]
        sub_labels = [article_labels[i] for i in problem_indices]
        
        sns.heatmap(sub_cm, annot=True, fmt='d', cmap='Blues', xticklabels=sub_labels, yticklabels=sub_labels)
        plt.title('Confusion Matrix - Focused Article Types')
    else:
        sns.heatmap(cm_article, annot=True, fmt='d', cmap='Blues', xticklabels=article_labels, yticklabels=article_labels)
        plt.title('Confusion Matrix - Article Types')
    
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.tight_layout()
    plt.savefig('confusion_matrix.png')
    plt.show()
    
    # Print Classification Report
    print("\nClassification Report - Gender:")
    print(classification_report(y_true_gender, y_pred_gender, target_names=gender_labels))
    
    print("\nClassification Report - Usage:")
    print(classification_report(y_true_usage, y_pred_usage, target_names=usage_labels))
    
    # Untuk artikel, mungkin terlalu banyak kelas, kita fokus pada kelas yang bermasalah
    problem_classes_idx = [i for i, label in enumerate(article_labels) 
                          if 'Heel' in label or 'Shirt' in label or 'Tshirt' in label or 'Shoes' in label]
    
    # FIX: Filtering both true and predicted values based on problematic indices
    filtered_indices = [i for i in range(len(y_true_article)) if y_true_article[i] in problem_classes_idx]
    y_true_filtered = [y_true_article[i] for i in filtered_indices]
    y_pred_filtered = [y_pred_article[i] for i in filtered_indices]
    
    # Get unique classes present in the filtered data
    present_classes = sorted(list(set(y_true_filtered + y_pred_filtered)))
    target_names = [article_labels[i] for i in present_classes]
    
    print("\nClassification Report - Selected Article Types:")
    try:
        report = classification_report(
            y_true_filtered,
            y_pred_filtered,
            labels=present_classes,  # Use the exact class indices present in data
            target_names=target_names,
            zero_division=0
        )
        print(report)
    except Exception as e:
        print(f"Error generating classification report for selected article types: {e}")
        print(f"Number of unique true classes: {len(set(y_true_filtered))}")
        print(f"Number of unique predicted classes: {len(set(y_pred_filtered))}")
        print(f"Total unique classes: {len(present_classes)}")
        print(f"Number of target_names: {len(target_names)}")
    
    # Analisis error pada kasus-kasus spesifik
    analyze_specific_issues(test_df, y_pred_gender, y_pred_usage, y_pred_article, 
                            gender_labels, usage_labels, article_labels)
    
    return (y_pred_gender, y_pred_usage, y_pred_article), cm_gender, cm_usage, cm_article
```

```python
def analyze_specific_issues(df, pred_gender, pred_usage, pred_article, gender_labels, usage_labels, article_labels):
    """Analisis mendalam untuk kasus-kasus spesifik yang bermasalah"""
    # Tambahkan prediksi ke dataframe
    df_copy = df.copy()
    df_copy['pred_gender'] = [gender_labels[i] for i in pred_gender]
    df_copy['pred_usage'] = [usage_labels[i] for i in pred_usage]
    df_copy['pred_article'] = [article_labels[i] for i in pred_article]
    
    # Tambahkan kolom true values
    df_copy['true_gender'] = [gender_labels[i] for i in df_copy['gender_encoded']]
    df_copy['true_usage'] = [usage_labels[i] for i in df_copy['usage_encoded']]
    df_copy['true_article'] = [article_labels[i] for i in df_copy['article_encoded']]
    
    # Tambahkan flag untuk prediksi yang salah
    df_copy['gender_correct'] = df_copy['true_gender'] == df_copy['pred_gender']
    df_copy['usage_correct'] = df_copy['true_usage'] == df_copy['pred_usage']
    df_copy['article_correct'] = df_copy['true_article'] == df_copy['pred_article']
    
    # Analisis untuk heels
    heels_df = df_copy[df_copy['true_article'].str.contains('Heel', case=False, na=False)]
    print(f"\n===== ANALISIS UNTUK HEELS (Total: {len(heels_df)}) =====")
    heels_incorrect = heels_df[~heels_df['article_correct']]
    if len(heels_incorrect) > 0:
        print(f"Total salah prediksi: {len(heels_incorrect)} ({len(heels_incorrect)/len(heels_df)*100:.2f}%)")
        wrong_predictions = heels_incorrect['pred_article'].value_counts()
        print("Salah prediksi sebagai:")
        print(wrong_predictions)
    else:
        print("Semua prediksi benar!")
    
    # Analisis untuk shirts dan tshirts wanita
    women_tops_df = df_copy[(df_copy['true_gender'] == 'Women') & 
                           (df_copy['true_article'].str.contains('Shirt|Tshirt', case=False, na=False))]
    print(f"\n===== ANALISIS UNTUK SHIRTS/TSHIRTS WANITA (Total: {len(women_tops_df)}) =====")
    women_tops_incorrect = women_tops_df[~women_tops_df['gender_correct']]
    if len(women_tops_incorrect) > 0:
        print(f"Total salah prediksi gender: {len(women_tops_incorrect)} ({len(women_tops_incorrect)/len(women_tops_df)*100:.2f}%)")
        wrong_gender = women_tops_incorrect['pred_gender'].value_counts()
        print("Salah prediksi gender sebagai:")
        print(wrong_gender)
    else:
        print("Semua prediksi gender benar!")
    
    # Analisis untuk shirts dan tshirts pria
    men_tops_df = df_copy[(df_copy['true_gender'] == 'Men') & 
                         (df_copy['true_article'].str.contains('Shirt|Tshirt', case=False, na=False))]
    print(f"\n===== ANALISIS UNTUK SHIRTS/TSHIRTS PRIA (Total: {len(men_tops_df)}) =====")
    men_tops_incorrect = men_tops_df[~men_tops_df['article_correct']]
    if len(men_tops_incorrect) > 0:
        print(f"Total salah prediksi artikel: {len(men_tops_incorrect)} ({len(men_tops_incorrect)/len(men_tops_df)*100:.2f}%)")
        wrong_article = men_tops_incorrect['pred_article'].value_counts()
        print("Salah prediksi artikel sebagai:")
        print(wrong_article)
    else:
        print("Semua prediksi artikel benar!")

# Load model dan evaluasi
try:
    # Coba load model dari saved file
    model = load_model('best_robust_model.h5')
    print("Model berhasil dimuat dari file.")
except:
    print("Model tidak ditemukan, gunakan model yang sudah dilatih.")
    model = robust_model  # Asumsikan model sudah ada di memori

# Evaluasi model
predictions, cm_gender, cm_usage, cm_article = evaluate_model_with_confusion_matrix(model, val_df, BATCH_SIZE)
```
